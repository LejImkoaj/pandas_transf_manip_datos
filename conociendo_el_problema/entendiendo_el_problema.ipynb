```cantidad_camas:``` informa el número de camas disponibles. ```=> Contiene una Lista Numerica```

```cuota_deposito:``` informa la tarifa mínima de depósito para la seguridad del hosting. ```=> Contiene una Lista Numerica```

```cuota_limpieza:``` informa el cargo cobrado por el servicio de limpieza. ```=> Contiene una Lista Numerica```

# 2.Datos Numericos

````python
columnas = list(datos.columns)
columnas
````

## Desglosamos las Columnas que son Listas

````python
datos_explode = datos.explode(columnas[3:]) # Explodeamos las columnas que contienen listas
datos_explode.reset_index(drop=True, inplace=True) # Reiniciamos el índice

datos.head(3)
````

## Aplicamos el .info() - Detalle del typo de Dato de las Columnas

Vemos que todas las Columnas estan de Dtype object.
Y no es asi

````python
datos_explode.info()
````

### Importamos la Biblioteca numerica de Numpy

````python
import numpy as np
````

### Convertimos la Columna a Tipo Numerico Entero

````python
datos_explode['max_hospedes'] = datos_explode['max_hospedes'].astype(np.int64) # Convertimos la columna 'max_hospedes' a tipo entero
                                                               # No Funciona si algun valor no es Entero
                                                           
````

### Cambiar el tipo de Dato a Varias Columnas

````python
col_numericas = ['cantidad_baños', 'cantidad_cuartos', 'cantidad_camas'] # listamos las columnas que van a ser convertidas a numéricas
````

````python
datos_explode[col_numericas] = datos_explode[col_numericas].astype(np.int64)  # Convertimos las columnas a tipo entero
datos_explode.info()
````

### Convertimos la tabla evaluacion_general en float - por que tiene un punto

````python
datos_explode['evaluacion_general'] = datos_explode['evaluacion_general'].astype(np.float64)  # Convertimos la columna a tipo float
datos_explode.info()
````

### Examinando la Tabla Precio - Ya que el numero tiene un signo de $

````python
datos_explode['precio'] = datos_explode['precio'].apply(lambda x: float(x.replace('$', '').replace(',', '').strip()))  # Limpiamos y convertimos la columna 'precio' a tipo float
````

````python
datos_explode.info()
````

### Examinando las tablas - cuota_deposito, cuota_limpiza
```Aplicando => map``` que es para DataFrames

````python
datos_explode[['cuota_deposito', 'cuota_limpieza']] = datos_explode[['cuota_deposito', 'cuota_limpieza']].map(lambda x: float(x.replace('$', '').replace(',', '').strip())) # Limpiamos y convertimos las columnas 'cuota_deposito' y 'cuota_limpieza' a tipo float
````

````python
datos_explode[['cuota_deposito', 'cuota_limpieza']] = datos_explode[['cuota_deposito', 'cuota_limpieza']].astype(np.float64) 
````

````python
datos_explode.info()
````
